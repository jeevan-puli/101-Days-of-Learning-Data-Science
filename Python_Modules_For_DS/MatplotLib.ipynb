# MatplotLib
- Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.
### Some of the major Pros of Matplotlib are:
- Generally easy to get started for simple plots.
- Support for custom labels and texts
- Great control of every element in a figure.
- High-quality output in many formats.
- Very customizable in general. 

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'StudyHours': [10,11.5,9,16,9.25,1,11.5,9,8.5,14.5,15.5,13.75,9,8,15.50],
                    'Grade': [50,50,47,97,49,3,53,42, 26, 74,82, 62,37,15, 165]})
df
```

```python
df.info()
```

## Analysis
- Univariate Analysis --> Analyzing single variable at a time.
- Bivariate Analysis --> Analyzing considering 2 Variables at a time.
- Multivariate Analysis --> Analyzing more than 2 variable ata a time. 

#### Univariate Analysis. 
##### Histogram
- In histograms, x axis contains a variable and y axis will be frequency of the variable.
###### Important Parameters of Histogram. 
- Data
- bin(default = 10) (width of each box).

```python
plt.hist(df['Grade'])
plt.xlabel("Garde")  #Assigning x axis name.
plt.ylabel("Count/Frequency") #Assigning y axis name.
plt.title("Histogram of Grade") #Assigning title of the histogram.
plt.show()
```

- the width of each box is bin size.
-  bin size = (max value - min value) / number of data points.

```python
plt.hist(df['Grade'],bins=20)
plt.xlabel("Garde")  #Assigning x axis name.
plt.ylabel("Count/Frequency") #Assigning y axis name.
plt.title("Histogram of Grade") #Assigning title of the histogram.
plt.show()
```

#### Box Plot
- Use to check if given data has outliers (concept from statistics)

```python
plt.boxplot(df["Grade"])
plt.title("Boxplot of Grades")
plt.show()
```

#### Bivariate Analysis
##### Scotter Plot
- Marking the data points on the graph.
- Use to check::
- - linearity.
  - direction.
  - Strength.
##### Important Parameters in Scatterplot. 
- X-axis values
- Y-axis Values.
- Color (default = blue)
- Marker(default = "o")
- Marker size (s)

```python
x = df["StudyHours"]
y = df["Grade"]

plt.scatter(x,y,color="blue",marker="o",s=30)
plt.show()
```

```python
# Same as above with labels.
x = df["StudyHours"]
y = df["Grade"]

plt.scatter(x,y,color="blue",marker="o",s=30)
plt.xlabel("Grades")  #Assigning x axis name.
plt.ylabel("StudyHours") #Assigning y axis name.
plt.title("Histogram of Grade") #Assign the title of histgram.
plt.show()
```

```python
# If you want to take grade on y axis and studyhours on x.

x = df["StudyHours"]
y = df["Grade"]

plt.scatter(y,x,color="blue",marker="o",s=30)
plt.xlabel("StudyHours")  #Assigning x axis name.
plt.ylabel("Grade") #Assigning y axis name.
plt.title("Histogram of Grade")
plt.show()
```

```python
# if you want to change the colour, marker and size.

x = df["StudyHours"]
y = df["Grade"]

plt.scatter(y,x,color="red",marker="*",s=40)
plt.xlabel("StudyHours")  #Assigning x axis name.
plt.ylabel("Grade") #Assigning y axis name.
plt.title("Histogram of Grades vs Studyhours")
plt.show()
```

#### Line Plot
- Use:: Represents the time series data.
- More useful if x axis is in proper order. (Asending or decending). 

```python
x = df["StudyHours"]
y = df["Grade"]

plt.plot(x,y)
plt.show()
```

```python
x = df["StudyHours"]
y = df["Grade"]

plt.plot(y,x)
plt.show()
```

```python
plt.plot(x,color='green',linewidth=3.0)
plt.show()

# if we do not provide two variables it will take by default index of row. 
```

```python
a = np.arange(6)
b = a
c = a**2
d = a**3
```

```python
plt.plot(a,b)
plt.show()
```

```python
plt.plot(a,c)
plt.show()
```

```python
plt.plot(a,d)
plt.show()
```

```python
# To show all plots in one line. 
plt.plot(a,b)
plt.plot(a,c)
plt.plot(a,d)
plt.show()
```

```python
# To plot with legends and it is only possible is all the same x axis. 
plt.plot(a,b, label="linear equation")
plt.plot(a,c,label="quadritc equation")
plt.plot(a,d, label="cubic equation") #assigning lable
plt.legend() #Asking to show legend. 
plt.show() 
```
