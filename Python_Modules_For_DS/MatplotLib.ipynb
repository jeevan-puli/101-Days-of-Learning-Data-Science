## Analysis
- Univariate Analysis --> Analyzing single variable at a time.
- Bivariate Analysis --> Analyzing considering 2 Variables at a time.
- Multivariate Analysis --> Analyzing more than 2 variable ata a time. 

#### Univariate Analysis. 
##### Histogram
- In histograms, x axis contains a variable and y axis will be frequency of the variable.
###### Important Parameters of Histogram. 
- Data
- bin(default = 10) (width of each box).

```python
plt.hist(df['Grade'])
plt.xlabel("Garde")  #Assigning x axis name.
plt.ylabel("Count/Frequency") #Assigning y axis name.
plt.title("Histogram of Grade") #Assigning title of the histogram.
plt.show()
```

- the width of each box is bin size.
-  bin size = (max value - min value) / number of data points.

```python
plt.hist(df['Grade'],bins=20)
plt.xlabel("Garde")  #Assigning x axis name.
plt.ylabel("Count/Frequency") #Assigning y axis name.
plt.title("Histogram of Grade") #Assigning title of the histogram.
plt.show()
```

#### Box Plot
- Use to check if given data has outliers (concept from statistics)

```python
plt.boxplot(df["Grade"])
plt.title("Boxplot of Grades")
plt.show()
```

#### Bivariate Analysis
##### Scotter Plot
- Marking the data points on the graph.
- Use to check::
- - linearity.
  - direction.
  - Strength.
##### Important Parameters in Scatterplot. 
- X-axis values
- Y-axis Values.
- Color (default = blue)
- Marker(default = "o")
- Marker size (s)

```python
x = df["StudyHours"]
y = df["Grade"]

plt.scatter(x,y,color="blue",marker="o",s=30)
plt.show()
```

```python
# Same as above with labels.
x = df["StudyHours"]
y = df["Grade"]

plt.scatter(x,y,color="blue",marker="o",s=30)
plt.xlabel("Grades")  #Assigning x axis name.
plt.ylabel("StudyHours") #Assigning y axis name.
plt.title("Histogram of Grade") #Assign the title of histgram.
plt.show()
```

```python
# If you want to take grade on y axis and studyhours on x.

x = df["StudyHours"]
y = df["Grade"]

plt.scatter(y,x,color="blue",marker="o",s=30)
plt.xlabel("StudyHours")  #Assigning x axis name.
plt.ylabel("Grade") #Assigning y axis name.
plt.title("Histogram of Grade")
plt.show()
```

```python
# if you want to change the colour, marker and size.

x = df["StudyHours"]
y = df["Grade"]

plt.scatter(y,x,color="red",marker="*",s=40)
plt.xlabel("StudyHours")  #Assigning x axis name.
plt.ylabel("Grade") #Assigning y axis name.
plt.title("Histogram of Grades vs Studyhours")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

days = [1,2,3,4,5,6,7]

max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

plt.plot(days, max_t, label = "max")
plt.plot(days, min_t, label = "min")
plt.plot(days, avg_t, label = "average")

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.legend()
plt.show()
```

#### Discreate Univariate Analyse. 

##### Pie Chart. 

```python
education = ["B.tech","M.tech","B.SC","M.sc","Phd"]
no_of_students = [140,5,30,8,12]
```

```python
plt.pie(no_of_students,labels=education,autopct="%1.1f%%",radius=1.5,explode=[0.1,0,0.1,0,0])
plt.show()
```

### Discreate Data Bar Plot.

```python
education = ["B.tech","M.tech","B.SC","M.sc","Phd"]
no_of_students = [140,5,30,8,12]
# Vertical bar chart. 
plt.bar(education,no_of_students)
plt.ylabel("no_of_students")
plt.title("Students in Groups")
plt.show()
```

```python
#horizontal bar chart. 
education = ["B.tech","M.tech","B.SC","M.sc","Phd"]
no_of_students = [140,5,30,8,12]

plt.barh(education,no_of_students)
plt.ylabel("no_of_students")
plt.title("Students in Groups")
plt.show()
```

### Stacked Bar Plot. 

```python
company = ["Google","Amazon","Microsoft","Meta"]
revenue = [90,136,89,27]
profit = [40,20,34,12]

plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")
plt.legend()
plt.show()
```

```python
company = ["Google","Amazon","Microsoft","Meta"]
revenue = [90,136,89,27]
profit = [40,20,34,12]

xpos = np.arange(len(company))
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Proifts")

plt.xticks(xpos,company)
plt.ylabel("Revenue(bln)")
plt.title("US Technology Stocks")
plt.legend()
plt.show()
```

### Subplots

```python
#Program for drawing multiple bar charts on one image.
ecommerce =['Snapdeal','Alibaba','Amazon','Flipkart']
Q1_Profit = [45,100,70,40]
Q2_Profit = [40,105, 65, 45]
Q3_Profit =[42, 120, 72, 50]
Q4_Profit =[34, 115, 60, 48]

#Creating different bar charts on one image using subplot () function.
plt. figure(figsize=(10,10))

#Creating bar chart in first cell of figure having 3 rows, 2 columns. 
plt.subplot (3,2,1)
plt.bar (ecommerce, Q1_Profit)
plt.title('Quarter1 Profit')

#Creating a bar chart in second cell. 
plt.subplot (3,2,2)
plt.bar (ecommerce, Q2_Profit)
plt.title('Quarter2 Profit')

#Creating a bar chart in third cell.
plt.subplot(3,2,3)
plt.bar(ecommerce, Q3_Profit)
plt.title('Quarter3 Profit')

#Creating a bar chart in sixth cell.
plt.subplot(3,2,6)
plt.bar(ecommerce, Q4_Profit)
plt.title('Quarter3 Profit')
```
