### Histogram/Distribution plot. 

```python
df = pd.read_csv("tips.csv")
df.head()
```

```python
sns.displot(df["total_bill"],bins=10)  #bin will give no of bars.
plt.show()
```

```python
sns.displot(df["total_bill"],bins=10,kde=True)  #kde(Kernal density estimation) will draw denisty curve 
plt.show()
```

### Inferences (conclusion, observation) from above:: The total_bill is right skewed. 

#### ++--------------------------------------------------------------------------------++

### Box Plot

```python
sns.boxplot(y=df["tip"])  # to put tip values on y axis. 
plt.show()
```

```python
sns.boxplot(x=df["tip"])  # to put tip values on x axis. 
plt.show()
```

### Inference:: This plot shows that there are outliers in the tip column. 

#### ++--------------------------------------------------------------------------------++

### Scatter Plot

```python
sns.scatterplot(x=df["total_bill"],y=df["tip"])  #Option 1.
plt.show()
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"])  #Option 2,
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data = df)  #Option 3. there data = will refere which data frame should the code consider. In this assigned df as name to my data frame. 
plt.show()
```

```python
sns.scatterplot(x="total_bill",y="tip",data = df)  #Option 4. there data = will refere which data frame should the code consider. In this assigned df as name to my data frame. 
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",hue = "sex",data = df)  #hue will help distingush plot points based on another variable and this is done with different colors. Here want to know who paid the time (male or female) so I used "sex" as hue.
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",style = "sex",data = df)  #hue will help distingush plot points based on another variable and this is done with different style plot points. Here want to know who paid the time (male or female) so I used "sex" as hue.
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",size = "sex",data = df)  #hue will help distingush plot points based on another variable(only discreate column) and this is done with different size plot points. Here want to know who paid the time (male or female) so I used "sex" as hue.
plt.show()
```

### Inference:: From above plot we can tell that, tip amout increase with total bill value and number of bills paid by male are higher than female.
