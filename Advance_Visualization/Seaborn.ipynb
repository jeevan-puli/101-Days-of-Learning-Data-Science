### Histogram/Distribution plot. 

```python
df = pd.read_csv("tips.csv")
df.head()
```

```python
sns.displot(df["total_bill"],bins=10)  #bin will give no of bars.
plt.show()
```

```python
sns.displot(df["total_bill"],bins=10,kde=True)  #kde(Kernal density estimation) will draw denisty curve 
plt.show()
```

### Inferences (conclusion, observation) from above:: The total_bill is right skewed. 

#### ++--------------------------------------------------------------------------------++

### Box Plot

```python
sns.boxplot(y=df["tip"])  # to put tip values on y axis. 
plt.show()
```

```python
sns.boxplot(x=df["tip"])  # to put tip values on x axis. 
plt.show()
```

### Inference:: This plot shows that there are outliers in the tip column.

## Join Plot
- A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation.
- It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot. 

```python
sns.jointplot(x="total_bill",y="tip",data=df)
plt.show()
```

 #### ++--------------------------------------------------------------------------------++

## Violin Plot.
- Violine plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot. 

```python
sns.violinplot(x = "total_bill", y = "day", data = df)
plt.show()
```

 #### ++--------------------------------------------------------------------------------++

 ## Pair Plot - Multiple continuous variables. 
 - A "Pairs Plot" is also known as a scatterplot, in which one variable in the same data row is matched iwth another variables value.
 - Plair plots are just elaborations on this showing all variables paired with all the other variables. 

```python
sns.pairplot(df,vars=continous)
plt.show()
```

 #### ++--------------------------------------------------------------------------------++

 ## Heat Map
 -  A heatmap uses colored cells to represent relation between variables.

 ## Heatmap (for correlation)
 - A correlation heatmap uses colored cells to show a @D correlation matrix(table) between two numeric dimensions.
 - It is very important in Feature Selection.

   Correlation(r) = COV(x,y) / σxσy

```python
c_m = df[["total_bill","tip"]].corr()   # I am first finding the correlation between two variables. 
c_m
```

```python
sns.heatmap(c_m,annot=True)  # I am ploting the heatmap for above correlation table. 
plt.show()
```

## Count Plot. 

```python
df["smoker"].value_counts()
```

```python
sns.countplot(x=df["smoker"])  # This will produce graph on x - axis. 

plt.savefig("CO")  # This will save the file to the working directory. 
plt.show()
```

## Bar Plot

```python
sns.catplot(x = "sex", y = "total_bill",data = df) # by default it will plot the point. 

plt.show()
```

```python
sns.catplot(x = "sex", y = "total_bill",data = df, kind = "bar") # this will produce bar plot for same data. The black line indicate that there are multiple bars are over laping. 

plt.show()
```

```python
sns.catplot(x = "day", y = "total_bill",data = df, kind = "bar", hue = "sex") # this will produce bar plot for same data. The black line indicate that there are multiple bars are over laping. 

plt.show()
```

```python
sns.catplot(x = "smoker", y = "total_bill",data = df, kind = "bar", hue = "size") # this will produce bar plot for same data. The black line indicate that there are multiple bars are over laping. 

plt.show()
```
