# AdaBoost
- AdaBoost (Adaptive Boosting) works by using an ensemble of weak learners and then combining them through the use of weighted sum.
- Adaboost adapts by using preiously created weak learners in order to adjust misclassified instances for the next created weak learner.

#### What is a weak learner?
- A weak model is a model that is too simple to predict output on its own.
- The weakest decision tree posible would be a stump, one mode and two leaves.

## Step 1: Business Problem.
- Develop a predictive model to identify whether mushroom is good or poisonous based on various features of mushrooms.
- A general guideline for features, people should look out for when picking mushrooms.

# 3.2 Data Cleaning.

```python
df.isna().sum()
```

## 3.3 Encoding.

```python
categorical_data = df.drop("class",axis=1)
```

```python
# Set make input data frame and output
X = pd.get_dummies(categorical_data, drop_first = True)
y = df["class"] 
```

```python
X.head()
```

## 3.4 Train / Test Split.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

## Step 4: Modelling (Base Line Model)

```python
# Adaboost Classifier Model with default Parameters
# 4.1 Modelling
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier()
model.fit(X_train,y_train)

# 4.2 Prediction
ypred_train = model.predict(X_train)
ypred_test = model.predict(X_test)
```

## Step 5: Evaluation

```python
# 5.1 Train & Test Accuracy.
from sklearn.metrics import accuracy_score
print("Train Accuracy::" , accuracy_score(y_train,ypred_train))
print("Test accuracy:", accuracy_score(y_test,ypred_test))

# 5.2 Cross Validation Scoer of model
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,X,y,cv=5)
print("Cross Validation score:", scores.mean())
```

- Even though train and test accuracys are good, the cross validaton score is not equal are near the test / train accuracy. So this is not a good model. 

## Step 6: Find Important Features. 

```python
model.feature_importances_
```

```python
feats = pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=["Importance"])
feats
```

```python
important_features = feats[feats["Importance"]>0]
important_features
```

```python
feats[feats["Importance"]>0].index
```

```python
imp_features_list = feats[feats["Importance"]>0].index.to_list()
imp_features_list
```

## Step 7: Hyperparameter Tuning.

```python
from sklearn.model_selection import GridSearchCV

estimator = AdaBoostClassifier()

param_grid = {"n_estimators":list(range(1,101))}

grid = GridSearchCV(estimator, param_grid, cv=5, scoring="accuracy")

grid.fit(X_train,y_train)

grid.best_params_
```

## Step 8: Model Rebuild with best Hyperparameters and important features. 

```python
# Drop unimportant features and create new X data frame. 
X = X[imp_features_list]

y = df["class"]

# Train / Test Split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=101)

#modelling

final_model = AdaBoostClassifier(n_estimators=20)
final_model.fit(X_train,y_train)

#predictions
ypredf_train = final_model.predict(X_train)
ypredf_test = final_model.predict(X_test)

#Evaluation
from sklearn.metrics import accuracy_score
print("Train accuracy:", accuracy_score(ypredf_train,y_train))
print("Test accuracy:", accuracy_score(ypredf_test,y_test))

from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,X,y,cv=5)
print("Cross Validation score:", scores.mean())
```

## Step 9: Inferences.

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,ypredf_test)
```

```python
#Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test,ypredf_test))
```
