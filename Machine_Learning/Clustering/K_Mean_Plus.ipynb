# Clustering

- A group of objects that are similar to other objects in the cluster and dissimilar to data points in other clusters.

### Applications:
- Retail / Marketing
- Banking.
- Insurance.

### Different clustering algorithms
- Partitioned-based clustering
- Hierarchical Clustering
- Density-based Clustering.

# ------------------

### Partition based (K means)
- Step 1: Choose the number K of clusters.
- Step 2: Select at random K points, the centroids (Not necessarily from your dataset)
- Step 3: Assign each data point to the closest centroid --> That forms K clusters
- Step 4: Compute and place the new centroid of each cluster.
- Step 5: Reassign each data point to the new closest centroid.If any reassignment took place, go to setp 4 other wise go to FIN.

### The Random Initialisation Trap.
- One major drawback of K-Mean clustering is the random initialisation of centroids. The formation of clusters is closely bound by the initial position of a centroid. The random positioning of the centroids can completely alter clusters and can result in a random formation.

The solution is K-mean++ is an algorithm that is used to initialise the k-Means algorithm. 

```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset["Gender"] = le.fit_transform(dataset["Gender"])
dataset
```

```python
X = dataset.drop("CustomerID",axis=1).values
```

```python
X
```

# Modelling

```python
from sklearn.cluster import KMeans
```

```python
# using the elbow method to find the optimal number of clusters. 
WCSS = []

for k in range(1,11):
    kmeans = KMeans(n_clusters = k, init = "k-means++")
    kmeans.fit(X)
    WCSS.append(kmeans.inertia_)
    
WCSS
```

```python
## Elbow Curve
import matplotlib.pyplot as plt
plt.plot(range(1,11),WCSS)

plt.xticks(range(1,11))
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")

plt.show()
```

## Fit the K-Mean model on the data. 

```python
#Model
kmeans = KMeans(n_clusters = 5, init = "k-means++", random_state = 42)
```

```python
#predict
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
y_pred = pd.DataFrame(y_kmeans,columns=["Cluster"])
y_pred
```

```python
df = pd.concat([dataset,y_pred],axis=1)
df
```

```python
df[df["Cluster"] == 3]
```

```python
# lets see the scatter plot. 
x = df["Annual Income (k$)"]
y = df["Spending Score (1-100)"]

plt.scatter(x,y)
```

```python
X = dataset.iloc[:,[3,4]].values

kmeans = KMeans(n_clusters = 5, init = "k-means++", random_state = 42)

y_kmeans = kmeans.fit_predict(X)
```

```python
# lets see the clustering in scatter plot. 
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1],
             s = 100, c = 'red', label = 'Cluster 1')


plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1],
            s = 100, c = 'blue', label = 'Cluster 2')


plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1],
            s = 100, c = 'green', label = 'Cluster 3')


plt. scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1],
             s = 100, c = 'cyan', label = 'Cluster 4')

plt. scatter (X[y_kmeans == 4, 0], X[y_kmeans == 4, 1],
              s = 100, c = 'magenta', label = 'Cluster 5')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s = 300, c = 'yellow', label = 'Centroids')

plt.title('Clusters of customers')

plt.xlabel('Annual Income (k$)')

plt.ylabel('Spending Score (1-100)')

plt.legend()

plt.show()
```
